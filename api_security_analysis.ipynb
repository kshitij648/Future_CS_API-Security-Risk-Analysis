```python
# ============================================
# API SECURITY RISK ANALYSIS TOOL
# Internship Task – Future Interns
# Author: Kshitij Pandey
# ============================================

# Install required libraries
import sys
import subprocess

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

try:
    import requests
except:
    install("requests")
    import requests

try:
    import pandas as pd
except:
    install("pandas")
    import pandas as pd


print("\n===============================")
print(" API SECURITY RISK ANALYZER ")
print("===============================\n")

# ---------------------------------
# API CONFIGURATION
# ---------------------------------

BASE_URL = "https://jsonplaceholder.typicode.com"

ENDPOINTS = [
    "/posts",
    "/users",
    "/comments",
    "/todos"
]

results = []

# ---------------------------------
# FUNCTION: TEST ENDPOINTS
# ---------------------------------

def test_endpoint(endpoint):

    url = BASE_URL + endpoint

    print(f"\nTesting Endpoint: {url}")

    try:

        response = requests.get(url)

        status = response.status_code
        headers = response.headers
        size = len(response.text)

        print("Status Code:", status)
        print("Response Size:", size)

        auth = "No Authentication Required"

        results.append({
            "Endpoint": endpoint,
            "Status Code": status,
            "Authentication": auth,
            "Data Size": size
        })

    except Exception as e:
        print("Error:", e)


# ---------------------------------
# RUN ENDPOINT SCAN
# ---------------------------------

print("\nScanning API Endpoints...\n")

for ep in ENDPOINTS:
    test_endpoint(ep)

print("\nEndpoint Scan Complete")


# ---------------------------------
# HEADER SECURITY CHECK
# ---------------------------------

print("\n===============================")
print(" HEADER SECURITY ANALYSIS ")
print("===============================\n")

url = BASE_URL + "/posts"

response = requests.get(url)

headers = response.headers

security_headers = [
    "X-Frame-Options",
    "X-Content-Type-Options",
    "Content-Security-Policy"
]

missing_headers = []

for header in security_headers:

    if header not in headers:
        missing_headers.append(header)

print("Missing Security Headers:", missing_headers)


# ---------------------------------
# DATA EXPOSURE TEST
# ---------------------------------

print("\n===============================")
print(" DATA EXPOSURE TEST ")
print("===============================\n")

response = requests.get(BASE_URL + "/users")

data = response.json()

print("Fields Returned By API:\n")

for field in data[0].keys():
    print("-", field)

print("\nPossible Risk: Excessive Data Exposure")


# ---------------------------------
# OBJECT LEVEL AUTHORIZATION TEST
# ---------------------------------

print("\n===============================")
print(" OBJECT LEVEL AUTHORIZATION TEST ")
print("===============================\n")

for i in range(1,6):

    url = BASE_URL + "/posts/" + str(i)

    r = requests.get(url)

    print("Accessing:", url)

    if r.status_code == 200:
        print("Data Accessible")
    else:
        print("Access Blocked")

print("\nPossible Risk: Broken Object Level Authorization")


# ---------------------------------
# POST REQUEST TEST
# ---------------------------------

print("\n===============================")
print(" INPUT VALIDATION TEST ")
print("===============================\n")

payload = {
    "title": "security test",
    "body": "api testing",
    "userId": 1
}

response = requests.post(BASE_URL + "/posts", json=payload)

print("POST Request Status:", response.status_code)
print("Response:", response.json())

print("\nPossible Risk: API accepts data without authentication")


# ---------------------------------
# CREATE RISK TABLE
# ---------------------------------

print("\n===============================")
print(" SECURITY RISK SUMMARY ")
print("===============================\n")

risk_data = [

["Unauthenticated API Access","Medium","Anyone can access API endpoints","Implement JWT/API Key authentication"],

["Excessive Data Exposure","Medium","API returns too many user fields","Limit returned fields"],

["Broken Object Level Authorization","High","Users can access other records","Validate user permissions"],

["Missing Rate Limiting","High","API vulnerable to request flooding","Add rate limiting"],

["Missing Security Headers","Low","Browser security headers missing","Add security headers"]

]

df = pd.DataFrame(risk_data,
columns=["Risk","Severity","Business Impact","Remediation"])

print(df)


# ---------------------------------
# GENERATE SECURITY REPORT
# ---------------------------------

print("\nGenerating Security Report...\n")

report = f"""

API SECURITY RISK ANALYSIS REPORT

API Tested:
JSONPlaceholder Demo API

Tools Used:
Jupyter Notebook
Python Requests Library

Methodology:
- API endpoint testing
- Header inspection
- Authentication analysis
- Data exposure testing

Findings:

1. Unauthenticated API Access
2. Excessive Data Exposure
3. Broken Object Level Authorization
4. Missing Rate Limiting
5. Missing Security Headers

Conclusion:
The API allows open access to multiple endpoints and lacks proper authentication and security headers.

Recommendation:
Implement authentication mechanisms, apply rate limiting, and validate user authorization.

"""

file = open("API_Security_Report.txt","w")
file.write(report)
file.close()

print("Report Saved As: API_Security_Report.txt")


print("\n===============================")
print(" ANALYSIS COMPLETE ")
print("===============================\n")
```
